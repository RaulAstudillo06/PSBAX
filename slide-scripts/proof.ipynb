```python
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
sns.set_theme(context='notebook', style='whitegrid', palette='deep', font='sans-serif', font_scale=1, color_codes=True, rc=None)
sns.set_style('whitegrid',{'axes.grid': False})


```

```python
fig, ax = plt.subplots(figsize=(10, 5))


def f4(x):
    if x > 0 and x < 10:
        return 5
    return 5 + 0.8 * (1 - np.abs(x - 5) / 5)

def f1(x):
    if x > 0 and x < 10:
        return 5
    else:
        return 5 + 2 * np.sin(np.pi * x / 10)

def f2(x):
    if x > 0 and x < 10:
        return 5
    else:
        return 5 + 4 * (1 - np.abs(x - 5) / 5)

def f3(x):
    if x > 0 and x < 10:
        return 5
    else:
        return 0.56


def f5(x):
    if x > 0 and x < 10:
        return 5
    if x < 5:
        return 5 + 4 * np.sin(np.pi * x / 20)
    else:
        return 5 + 4 * np.sin(np.pi * (10 - x) / 20)


x = np.linspace(-5, 15, 1000)

y1 = np.array([f1(xi) for xi in x])
y2 = np.array([f2(xi) for xi in x])
y3 = np.array([f3(xi) for xi in x])
y4 = np.array([f4(xi) for xi in x])
y5 = np.array([f5(xi) for xi in x])

plt.plot(x, y1, label='f1')
plt.plot(x, y2, label='f2')
plt.plot(x, y3, label='f3')
plt.plot(x, y4, label='f4')
plt.plot(x, y5, label='f5')


# plt.axvline(x=0, color='k', linestyle='--', alpha=0.5)
# plt.axvline(x=10, color='k', linestyle='--', alpha=0.5)
plt.axhline(y=5, color='k', linestyle='--', alpha=0.5)

plt.xlim(-5, 15)
plt.ylim(0, 10)
# plt.xlabel('x')
# plt.ylabel('f(x)')
# plt.title('5 Concave Down Functions Meeting at f(0) = f(10) = 5')
# plt.legend()
# plt.grid(True, alpha=0.3)

plt.tight_layout()
plt.savefig('figs/proof.svg')
plt.show()
```
